### 커피베이 메뉴 크롤링

https://www.coffeebay.com/home/menu/menu_

coffee, franobe, smoobella, etc 가 돌아가면서 페이지를 구성함

```python
# 제품 코드 추출
from bs4 import BeautifulSoup
from selenium import webdriver
from pprint import pprint
```

```python
# webdriver 실행
driver = webdriver.Chrome()

# target page 접근
driver.get("https://www.coffeebay.com/home/menu/menu_coffee")
```

```python
# html source 추출
html_source = driver.page_source
print(html_source)
```

```python
# BS로 html parsing
soup = BeautifulSoup(html_source, 'html.parser')
print(soup)
```

```python
# 원하는 항목의데이터만 추출
products_info = soup.select('div.menu-child-info-box p.menu-child-title-item span.menu-child-name-title')
pprint(products_info)
```

```python
# 제품 이름, 코드 추출
product_name = [product.text for product in products_info]
pprint(product_name)
```

```python
ingredients_info_title = soup.select("div.menu-child-info-box table:nth-child(4) thead")
pprint(ingredients_info_title)
```

```python
ingredients_info_body = soup.select("div.menu-child-info-box table:nth-child(4) tbody")
pprint(ingredients_info_body)
```
