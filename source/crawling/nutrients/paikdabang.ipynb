### 빽다방 메뉴 크롤링

```python
# 제품 코드 추출
from bs4 import BeautifulSoup
from selenium import webdriver
from pprint import pprint
```

#### 커피(coffee) 추출

```python
# webdriver 실행
driver = webdriver.Chrome()

# target page 접근
driver.get("https://paikdabang.com/menu/menu_coffee/")
```

```python
# html source 추출
html_source = driver.page_source
print(html_source)
```

```python
# BS로 html parsing
soup = BeautifulSoup(html_source, 'html.parser')
print(soup)
```

```python
urls_h3 = []
urls_ingred = []
product_name = []
info_ingred2 = []

for i in range(37):
    i = i+1
    # 각 카테고리에 따른 링크 만들기
    url_h3 = '.menu_list.clear ul li:nth-child(' + str(i) + ') div.hover h3'
    url_ingred = '.menu_list.clear ul li:nth-child(' + str(i) + ') div.hover div ul li div'
    urls_h3.append(url_h3)
    urls_ingred.append(url_ingred)
    
    
    i = i-1
    
    # 상품명 가져오기
    products = soup.select(urls_h3[i])
    
    # 영양정보 가져오기
    ingredients = soup.select(urls_ingred[i])
    info_ingred = []
    
    for ingredient in ingredients:
        info_ingred.append(ingredient.string)
        
    info_ingred2.append([info_ingred])
        
    for product in products:
        product_name.append([str(i), product.get_text()])
```

```python
print(product_name)
```

```python
# 복제본 만들어 핸들링
final_info_ingred = []
final_info_ingred = info_ingred2

# 정보 미제공 품목 있는지 확인
for i in range(37):
    print(final_info_ingred[i][0])
```

```python
# 미제공 항목 모두 공백으로 처리
final_info_ingred[30][0] = ['카페인 (mg)', '', '칼로리 (kcal)', '', '나트륨 (mg)', '', '당류 (g)', '', '포화지방 (g)', '', '단백질 (g)', '']
final_info_ingred[31][0] = ['카페인 (mg)', '', '칼로리 (kcal)', '', '나트륨 (mg)', '', '당류 (g)', '', '포화지방 (g)', '', '단백질 (g)', '']
```

```python
# 정보 미제공 품목 있는지 확인
for i in range(37):
    print(final_info_ingred[i][0])
```

```python
# result를 담을 빈 List 생성
result = []

# 제품별로 for loop
for i in range(37):
    container = dict()
    category = '커피'
    menu = product_name[i][1]
        
    # 제품 영양정보
    caffeine = final_info_ingred[i][0][1]
    kcal = final_info_ingred[i][0][3]
    sodium = final_info_ingred[i][0][5]
    sugars = final_info_ingred[i][0][7]
    trans_FAT = final_info_ingred[i][0][9]
    protein = final_info_ingred[i][0][11]
        
    
    container['분류'] = category
    container['메뉴'] = menu
    container['칼로리 (kcal)'] = kcal
    container['당류 (g)'] = sugars
    container['단백질 (g)'] = protein
    container['나트륨 (mg)'] = sodium
    container['포화지방 (g)'] = trans_FAT
    container['카페인 (mg)'] = caffeine
    
    result.append(container)

pprint(result)
```
