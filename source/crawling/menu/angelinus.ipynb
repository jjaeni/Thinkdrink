#### 엔제리너스 메뉴 크롤링

```python
# 제품 코드 추출
from bs4 import BeautifulSoup
from selenium import webdriver
from pprint import pprint
```

```python
# 커피, snow, 음료 순으로 메뉴인덱스가 4, 2, 3으로 바뀜
url = [4, 2, 3]
print(url)

for i in range(len(url)):
    url[i] = "https://www.angelinus.com/Menu/Menu.asp?MenuIdx="+str(url[i])
    
print(url)
```

```python
# webdriver 실행
driver = webdriver.Chrome("C:/Users/wodms/Jupyter_python/chromedriver.exe")

# target page 접근
driver.get(url[0])
```

```python
# html source 추출
html_source = driver.page_source
print(html_source)
```

```python
# BS로 html parsing
soup = BeautifulSoup(html_source, 'html.parser')
print(soup)
```

```python
# 원하는 항목의데이터만 추출
# "span", {"class":"Python3"}
products = soup.find("div", id="menuList")
products2 = products.select("ul li a")


print(products2)
```
