```python
# 초기세팅 시작
from urllib.request import urlopen
from urllib.parse import quote_plus as qp   # 아스키코드로 변환
from bs4 import BeautifulSoup
from selenium import webdriver
import time   # 드라이버의 크롤링 시간 지정
import requests
import shutil

# 이미지 추출할 태그 링크 생성
baseurl = "https://www.instagram.com/explore/tags/"
tagurl = input("검색할 태그 입력: ")
url = baseurl + qp(tagurl)   # qp() 입력한 언어를 아스키코드로 변환
print(url)

# webdriver 사용
driver = webdriver.Chrome()
driver.get(url)

time.sleep(3)   # 딜레이 방지용

html = driver.page_source
soup = BeautifulSoup(html)


imglist = []   # 이미지 빈 리스트 생성


for i in range(0, 11):   # 스크롤 횟수 지정 루프
    
    # 태그 추출 확인
    insta = soup.select("._aabd._aa8k._aanf")   # 이미지 클래스명
    
    for i in insta:
        print("https://www.instagram.com" + i.a["href"])   # a 태그의 속성 href만 가져오기
        imgurl = i.select_one("._aagv").img["src"]   # 이미지가 속한 태그의 src 태그만 가져오기
        imglist.append(imgurl)   # 만들어준 빈 리스트에 이미지url 삽입
        imglist = list(set(imglist))
        
        html = driver.page_source
        soup = BeautifulSoup(html)
        insta = soup.select("._aabd._aa8k._aanf")
        
        
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")   # 이동하고 싶은 위치로 스크롤 이동
    time.sleep(5)
    

n = 1

for i in range(len(imglist)):
    # This is the image url.
    image_url = imglist[n]
    
    # Open the url image, set stream to True, this will return the stream content.
    resp = requests.get(image_url, stream = True)
    
    # Open a local file with wb(write binary) permission
    local_file = open("Foam Foam Youthberry Tea/img/" + tagurl + str(n) + ".jpg", "wb")
    
    # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
    resp.raw.decode_content = True
    
    # Copy the response stream raw data to local image file.
    shutil.copyfileobj(resp.raw, local_file)
    
    # Remove the image url response object.
    n += 1
    
    del resp

driver.close()
```
